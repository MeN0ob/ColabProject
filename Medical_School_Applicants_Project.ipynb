How does the distribution of medical school applicants vary across racial and ethnic groups in the 2024-2025 cycle, and which groups are the most and least represented across the U.S?

```python
#Keep only relevant race/ethnicity columns and state names
race_columns = ['Applicants by State of Legal Residence', 'Asian', 'Black or African American', 'Hispanic, Latino, or of Spanish Origin', 'White']
#Drop rows with missing data and reset index
df_subset = df[race_columns].dropna().reset_index(drop = True)
#Rename columns for cleaner display
df_subset = df_subset.rename(columns = {'Applicants by State of Legal Residence':'State'})
df_subset.head(16)
```

```python
from scipy.stats import f_oneway
#One-way ANOVA: compare means across race groups
f_stat, p_val = f_oneway(df_subset['Asian'], df_subset['Black or African American'], df_subset['Hispanic, Latino, or of Spanish Origin'], df_subset['White'])
print(f"F-statistic:{f_stat:.2f}")
print(f"P-value:{p_val:.4f}")
if p_val < 0.05:
  print("Statistically significant difference in applicant counts among the groups.")
else:
  print("No significant difference in applicant counts among the groups.")
```

```python
#Calculate total number of applicants per group
total_applicants = df_subset[['Asian', 'Black or African American', 'Hispanic, Latino, or of Spanish Origin', 'White']].sum()
#Plot total applicants by race
plt.figure(figsize = (8,5))
sns.barplot(x = total_applicants.index, y = total_applicants.values, palette = 'Set2')
plt.ylabel("Total Applicants")
plt.title("Total Medical School Applicants by Rce/Ethnicity (2024 - 2025)")
plt.xticks(rotation = 30)
plt.tight_layout()
plt.show()
```

```python
#Adding a new column for total applicants per state
df_subset['Total'] = df_subset[['Asian', 'Black or African American', 'Hispanic, Latino, or of Spanish Origin', 'White']].sum(axis = 1)
#Creating a new df with percents
df_percent = df_subset.copy()
race_cols = ['Asian', 'Black or African American', 'Hispanic, Latino, or of Spanish Origin', 'White']
#Converting each column into percents for that state
for col in race_cols:
  df_percent[col] = (df_subset[col] / df_subset['Total']) * 100
df_percent[['State'] + race_cols].head()
```

```python
import matplotlib.pyplot as plt

states_to_plot = ['California', 'Texas', 'New York', 'Florida', 'Maryland']
df_plot = df_percent[df_percent['State'].isin(states_to_plot)]

plt.figure(figsize = (12,6))
# Plotting grouped bar chart
x = range(len(states_to_plot))
width = 0.2
for i, race in enumerate(race_cols):
  plt.bar([p + width * i for p in x], df_plot[race], width, label = race)
plt.xticks([p + width * (len(race_cols) / 2) for p in x], df_plot['State'], rotation = 30)
plt.ylabel('Percentage of Applicants')
plt.title('Racial Disribution of Applicants Across Selected States (2024 - 2025)')
plt.legend()
plt.tight_layout()
plt.show()
```

California's applicants says that they're mostly Asian and I believe that's because California has a high Asian population. Besides California and Texas applicants across the other states are mostly white. While Black and Hispanic applicants are underrepresented across the states.
